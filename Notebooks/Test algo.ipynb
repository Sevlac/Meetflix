## NETTOYAGE

```python
# Récupérer les stopwords pour le français
stop_words = set(stopwords.words('french'))
```

```python
df_words['keywords'] = df_words['keywords'].apply(ast.literal_eval)
```

```python
# Fonction pour nettoyer les mots-clés
def clean_keywords(keywords):
    return [
        word.lower().strip()
        for word in keywords
        if word.isalpha() and word.lower() not in stop_words
    ]

df_words['keywords_cleaned'] = df_words['keywords'].apply(clean_keywords)
```

### Get_Dummies

```python
df_words['keywords_cleaned_str'] = df_words['keywords_cleaned'].apply(lambda x: '|'.join(x))
keywords_dummies = df_words['keywords_cleaned_str'].str.get_dummies(sep='|')
keywords_dummies
```

```python
# On comtpe le nbre de morts distincts
keyword_occurrences = keywords_dummies.sum().sort_values(ascending=False)
```
